```python
import pandas as pd
import numpy as np
import plotly.express as px #plotly.express(px):Data Visulaisation library
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data=pd.read_csv('Superstore.csv',encoding='latin-1')
#encoding=latin-1:Special Character ko properly read karne ke liye
data
#print(data.head())
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
#Date Conversion:Order Date aur Ship Date columns ko determine format karne ke liye
```

Adding New Date-Based Columns

```python
data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of Week']=data['Order Date'].dt.dayofweek

#order month:Order date se month extract karte hain
#order year:Order date se year extract karte hain
#order day of week:Order date se day of week extract karte hain
```

```python
data.head()
```

Monthly Sales Analysis

```python
sales_by_month=data.groupby('Order Month')['Sales'].sum().reset_index()
fig=px.line(sales_by_month,x='Order Month',y='Sales',title='Monthly Sales Analysis')
fig.show()
```

This graph shows that Month of November has the highest sales and February has the lowest sales.

Sales Analysis by Category

```python
sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
fig=px.pie(sales_by_category,values='Sales',names='Category',hole=0.5,color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category',title_font=dict(size=24))
fig.show()
```

This graph shows that technology sector has the highest sales


```python
sales_by_subcategory=data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig=px.bar(sales_by_subcategory,x='Sub-Category',y='Sales',color='Sub-Category',title='Sales Analysis by Sub-Category')
fig.show()
```

This graph shows that Phones have the highest sales and Fastners have the lowest sale

Monthly Profit Analysis

```python
profit_by_month=data.groupby('Order Month')['Profit'].sum().reset_index()
fig=px.line(profit_by_month,x='Order Month',y='Profit',title='Monthly Profit Analysis')
fig.show()
```

This graph shows that 12th month has the highest profit and 1st month has the lowest profit

Profit Analysis by Category

```python
profit_by_category=data.groupby('Category')['Profit'].sum().reset_index()
fig=px.pie(profit_by_category,values='Profit',names='Category',hole=0.5,color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category',title_font=dict(size=24))
fig.show()
```

This graph shows that  Tech has the highest profit

Profit Analysis by Sub Category

```python
profit_subcategory=data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig=px.bar(profit_subcategory,x='Sub-Category',y='Profit',color='Sub-Category',title='Profit Analysis by Sub-Category')
fig.show()
```

This graph shows that copiers have the highest profit and tables have lowest profit

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_palette=colors.qualitative.Pastel
fig=go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Sales'],name='Sales',marker_color=color_palette[5]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Profit'],name='Profit',marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Segment',xaxis_title='Segment',yaxis_title='Amount')
fig.show()
```

This Graph shows that consumers have the highest sales and profit.

Analysis sales-to-profit ratio

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales-to-Profit Ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales-to-Profit Ratio']])
'''fig=go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Sales'],name='Sales',marker_color=color_palette[5]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Profit'],name='Profit',marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Segment',xaxis_title='Segment',yaxis_title='Amount')
fig.show()'''
```

```python
fig=px.bar(sales_profit_by_segment,x='Segment',y='Sales-to-Profit Ratio')
fig.update_traces(textposition='inside')
fig.update_layout(title='Sales-to-Profit Ratio by Customer Segment',xaxis_title='Segment',yaxis_title='Sales-to-Profit Ratio')
fig.show()
```

This graph shows that consumer segment has the highest sales to profit ratio
